### Переменные

Переменная - именованная область памяти, предназначенная для храненния данных ("коробочка" для любых данных, нужна чтобы по имени коробочки получить эти данные).



Правила именования переменных:

    - имя должно начинаться с буквы или _ (не с цифры или знака)

    - имя может включать буквы, цифры, _, но не другие символы

    - регистрозависимость (a = 5 и A = 5 это две разные переменные)
    
    - имя не может быть ключевым словом (def, if, for, while, ilif, else, class, type ...)

В Python принято именование через snake_case (some_variable = 5)

```python
# определение (объявление, инициализация)

# ! Нельзя обявить и не проинициализировать переменную!
# number
# ...
# number = 3

my_number = 10

my_string = 'Hello world!'
```

```python
# Переопределение

number = 5
number = 8

print(number) # 8
```

```python
# Множественное объявление и инициализация

a = b = c = 10

x, y, z = 1, 2, 3
```

### Ввод / вывод

Встроенные функции для вывода - **print()**, **display()**. 

print() можно передавать бесчисленное количество аргументов, значения будут выводиться через пробел:
print(a, b, c, ..... n)

```python
# print(), display()

string = 'Hello world!'

print(string)
```

```python
first = 'Hello'
second = 'world'
third = '!'

print (first, second, third)
```

Встроенная функция для ввода с клавиатуры - **input()**.

В input() можно передать промпт (подпись) для пользователя.

Значение, которое ввёл пользователь, лучше сохранять в переменную чтобы потом с ним работать.

input() **ВСЕГДА** возвращает строку. Независимо от того, что введёт пользователь.

Чтобы посмотреть тип какого-то значения существует встроенная функция **type()**.

```python
# input()

variable = input()

print(variable, type(variable))
```

```python
city = input('Введите город: ')

print('Введённый город:', city)
```

### Простые типы данных

##### Числовые 

    - int (целочисленный): 1, 5, 5000, -49
    
    - float (вещественный, с плавающей точкой): 1.0, -1.25, -5e2
    
    - complex (комплексное число: [вещественная_часть]+[мнимая_часть]j): 1+2j


##### Текстовые (строковые) 

    - str: '...', "...", '''...''' 
    
    - f-строка: f'... {} ...'

##### Логические 

    - boolean: True (истина), False (ложь)


##### NoneType 

    - None - ничто

```python
# f-строка использует интерполяцию строк (прямая вставка литерала (значения) или переменной в строку)

a = 5
b = 10
c = a + b

f_string = f'{a} + {b} = {c}'

print(f_string)

print(f'a is {a}, b is {b}, result of a + b is {a + b}')
```
